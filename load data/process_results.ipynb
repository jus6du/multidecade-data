```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objs as go
```

```python
reg_parameters_days = pd.read_csv('processing/reg_parameters_days_of_week_2010-2014.csv', index_col = [0])
reg_parameters_hours = pd.read_csv('processing/reg_parameters_hours_of_week_2010-2014.csv', index_col = [0])
reg_parameters_temp = pd.read_csv('processing/reg_parameters_temp_2010-2014.csv', index_col = [0])
reg_parameters_trend = pd.read_csv('processing/reg_parameters_trend_2010-2014.csv', index_col = [0])
```

```python
artificial_demand = pd.read_csv('europe_demand_artificial_1980-2020.csv')
temperatures = pd.read_csv('original_data/europe_temperatures_1980-2020.csv')
```

```python
year_to_extract = 2020  # Change this to the year you want to extract
artificial_demand['time'] = pd.to_datetime(artificial_demand['time'])
temperatures['time'] = pd.to_datetime(temperatures['time'])
# Filter the DataFrame to include only the data for the desired year
year_demand = artificial_demand[artificial_demand['time'].dt.year == year_to_extract]
year_temperature = temperatures[temperatures['time'].dt.year == year_to_extract]
```

```python
country_to_extract = 'DK'
year_country_demand = year_demand[country_to_extract]
year_country_temperature = year_temperature[country_to_extract]
```

```python
country_to_extract = 'FR'
year_country_demand = year_demand[country_to_extract]
year_country_temperature = year_temperature[country_to_extract]

fig, ax1 = plt.subplots()

# Tracé de la température sur le premier axe
ax1.plot(year_country_temperature, 'r-')
ax1.set_xlabel('Temps')
ax1.set_ylabel('Température (°C)', color='r')

# Créer un deuxième axe pour la consommation électrique
ax2 = ax1.twinx()
ax2.plot(year_country_demand, 'b-')
ax2.set_ylabel('Consommation électrique (kWh)', color='b')

# Ajouter une légende
ax1.legend(['Température'], loc='upper left')
ax2.legend(['Consommation électrique'], loc='upper right')

# Afficher le tracé
plt.show()
```

```python
df = year_demand
# Convertir les timestamps en datetime si nécessaire
df['time'] = pd.to_datetime(df['time'])

# Définir la colonne de timestamp comme l'index
df.set_index('time', inplace=True)
# Agréger les données par semaine
df_weekly = df.resample('W').mean()

# Créer la figure et les traces pour chaque série temporelle
fig = go.Figure()

for col in df.columns:
    if col != 'timestamp':  # Ignorer la colonne timestamp
        fig.add_trace(go.Scatter(x=df_weekly.index, y=df_weekly[col], mode='lines', name=col))

# Définir les titres et les étiquettes d'axes
fig.update_layout(title='Séries temporelles lissées par semaine',
                  xaxis_title='Temps',
                  yaxis_title='Valeur')

# Afficher le tracé
fig.show()
```
