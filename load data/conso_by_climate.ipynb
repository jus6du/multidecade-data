```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objs as go
```

## Load Artificial Demand

```python
countries = ['FI', 'KR', 'EG', 'IT', 'PL', 'CL', 'VE']
dic_climate = {'FI': 'Continental, cold summer',
            'KR':'Continental, hot summer', 
            'IT': 'Temperate, hot summer', 
             'CL': 'Arid, cold',
            'EG': 'Arid, hot', 
            'VE': 'Tropical',
            'PL': 'Continental, warm summer'}
```

```python
fig = go.Figure()

for c in countries:
    demand = pd.read_csv(f'{c}_demand_artificial_2021-2022.csv')
    demand =demand.set_index('Unnamed: 0')  
    fig.add_trace(go.Scatter(x=demand.index, y=demand[c], mode='lines', name=dic_climate[c]))

fig.update_layout(title='Artificial Demand',
                  xaxis_title='Time',
                  yaxis_title='Power(MW)')
fig.show()
```
